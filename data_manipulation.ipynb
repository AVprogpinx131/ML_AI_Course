```python
df = df.rename(columns={
    "39": "Age",
    " 77516": "Salary",
    " 40": "Week Working Hours",
    " 13": "Day Working Hours",
    " State-gov": "Workclass",
    " Bachelors": "Education Level",
    " Never-married": "Marital Status",
    " Adm-clerical": "Occupation",
    " Not-in-family": "Relationship",
    " White": "Race",
    " Male": "Gender",
    " United-States": "Working Locations",
    " <=50K": "Income"
})

df
```

```python
# A histogram showing the age distribution.
plt.hist(df["Age"], bins=25, color="skyblue")
```

```python
# A bar chart that allows you to compare average weekly working hours across different occupations using matplotlib.
plt.figure(figsize=(30, 6))
plt.bar(df["Occupation"], df["Week Working Hours"], color="skyblue")
```

```python
# A bar chart that allows you to compare average weekly working hours across different occupations
plt.figure(figsize=(10, 6))
sns.heatmap(df.select_dtypes(include=["number"]).corr(), annot=True, cmap="coolwarm")
plt.show()
```

```python
# Other visualizations
plt.figure(figsize=(30, 6))
income_counts = df.groupby(["Occupation", "Income"]).size().unstack()
income_counts.plot(kind='bar', figsize=(30, 6))
plt.title('Income Distribution by Occupation')
plt.xlabel('Occupation')
plt.ylabel('Count')
plt.legend(title='Income', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()


plt.figure(figsize=(30, 6))
income_counts = df.groupby(["Gender", "Income"]).size().unstack()
income_counts.plot(kind='bar', figsize=(30, 6))
plt.title('Income Distribution by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Income', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

Aga üldiselt numbrilised tunnused (töötunnid ja vanus) ei mõjuta palka, ehk võib järeldada, et palka võib hoopis rohkem mõjutada näiteks haridustase. Kõige suuremas tasakaalus on töötunnid (päevas ja nädalas korrelatsiooniga **0.015**)
